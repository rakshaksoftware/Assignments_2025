```python
#Q1.1
for i in range(1,101):
    if not (i%3 == 0 or i%5 == 0):
        print(i,end = '')
    if i%3 == 0:
        print('Fizz',end = '')
    if i%5 == 0:
        print('Buzz',end = '')
    print(' ',end = '')
```

```python
#Q1.2
import math
res = []
a = int(input())
b = int(input())
for i in range(a,b+1):
    if i == 2:
        res.append(i)
        continue
    if i == 1 or i%2 == 0:
        continue
    
    cond = True
    for j in range(3,int(math.sqrt(i))+1,2):
        if i%j == 0:
            cond = False
            break
    if cond:
        res.append(i)

print(res)
```

```python
#Q2.1
import numpy as np
rg = np.random.default_rng()
arr1,arr2 = rg.integers(1,11,[3,3]),rg.integers(1,11,[3,3])
print(arr1,arr2,sep = '\n')
print(arr1 + arr2)
print(arr1*arr2)
print(arr1 @ arr2)
print(np.linalg.det(arr1))
```

```python
#Q2.2
arr1 = np.random.normal(50,5,1000)
print(np.mean(arr1))
print(np.median(arr1))
print(np.std(arr1))
print(np.var(arr1))
print(np.percentile(arr1,25))
print(np.percentile(arr1,75))
```

```python
#Q5.1
import datetime
base = datetime.date(2025, 1, 1)
time = [base + datetime.timedelta(days=x) for x in range(31)]
x = np.linspace(0,2*np.pi,31)
temp = 15 + 10*np.sin(x)
df = pd.DataFrame({"time" : time,"temp" : temp})
roll = df["temp"].rolling(7).mean()
print(df[df["temp"]> 20])
plt.plot(df[df["temp"]> 20]["time"],df[df["temp"]> 20]["temp"],marker = "+",markersize = 20)
plt.plot(time,temp)
plt.xticks(rotation = 90)
plt.plot(time,roll,linestyle = '-.')
plt.legend(["greater than 20","data","average"])
plt.show()
```
